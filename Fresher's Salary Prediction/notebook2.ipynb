Dropping Columns

```python
df.drop(['ID','DOB','10board','12graduation','12board','CollegeID','CollegeCityID','CollegeCityTier','CollegeState','GraduationYear'], axis=1, inplace=True)
df.head()
```

```python
df = df.drop_duplicates()
```

Working on Specialization column

```python
df.Specialization.unique()
```

```python
specialization = df.Specialization.value_counts(ascending=False)
```

```python
specialization_lessthan10 = specialization[specialization<=10]
specialization_lessthan10
```

```python
def handelSpecialization(value):
    if value in specialization_lessthan10:
        return 'other'
    else:
        return value
df.Specialization = df.Specialization.apply(handelSpecialization)
df.Specialization.unique()
```
