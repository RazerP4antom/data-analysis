# Data Cleaning 

Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
df = pd.read_csv('Engineering_graduate_salary.csv')
df.head()
```

```python
df.columns
```

Dropping Columns

```python
df.drop(['ID','DOB','10board','12graduation','12board','CollegeID','CollegeCityID','CollegeCityTier','CollegeState','GraduationYear'], axis=1, inplace=True)
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Degree.unique()
```

Working on Specialization column

```python
df.Specialization.unique()
```

```python
specialization = df.Specialization.value_counts(ascending=False)
```

```python
specialization_lessthan10 = specialization[specialization<=10]
specialization_lessthan10
```

```python
def handelSpecialization(value):
    if value in specialization_lessthan10:
        return 'other'
    else:
        return value
df.Specialization = df.Specialization.apply(handelSpecialization)
df.Specialization.unique()
```

```python
df.head()
```

### Working on GPA Column

```python
plt.scatter(df.index, df['collegeGPA'])
```

```python
# Considering only GPA above 40
df = df[(df['collegeGPA'] > 40)]
df.shape
```

```python
# replacing -1 with nan 
df = df.replace(-1, np.nan)
```

```python
# filling null values with mean
col_with_nan = [column for column in df.columns if df.isna().sum()[column] > 0]
for column in col_with_nan:
    df[column] = df[column].fillna(df[column].mean())
```

### Data Preprocessing

```python
# since we cant pass string values to our model we must convert them into integers using Label Encoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df.Gender = le.fit_transform(df.Gender)
df.Degree = le.fit_transform(df.Degree)
df.Specialization = le.fit_transform(df.Specialization)

df.head()
```

### Creating XGBoost Model

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#StandardScaler since it's a regression problem so we scale down our data
from xgboost import XGBRegressor
```

```python
X = df.drop('Salary', axis=1)
y = df['Salary']

sc = StandardScaler()

X = sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True,random_state=0)
```

```python
xgb = XGBRegressor()

xgb.fit(X_train, y_train)
```

### Predicting test_X data

```python
predictions = xgb.predict(X_test)
```

```python
diff = y_test - predictions
```

### Visualising Difference

```python
plt.figure(figsize = (12,10))
sns.displot(diff)
```

### R^2 Score

```python
## checking R2 score

xgb_r2_score = xgb.score(X_test, y_test)

print("XGBoost R2 Score: ", (xgb_r2_score))
```
